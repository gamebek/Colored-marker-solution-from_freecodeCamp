```python
import numpy as np

numbers_list = [10, 20, 30, 40, 50]

# 1. Convert the list to a NumPy array for fast numerical operations
numbers_array = np.array(numbers_list)

# 2. Calculate statistics using NumPy functions
mean_value = np.mean(numbers_array)
max_value = np.max(numbers_array)
sum_value = np.sum(numbers_array)

# --- Output ---
print(f"numbers: {numbers_array}")
print(f"mean: {mean_value}")
print(f"max: {max_value}")
print(f"sum: {sum_value}")
```

1. ML Problem (1 sentence)

Predict how many loaves of bread a grocery store should bake each day.

2. ML Type

Supervised Learning
Because we have historical sales data (input) and the number of breads sold (output).
The model learns from labeled examples to predict future demand.

3. ML Workflow

Gather daily sales records.

Clean missing or incorrect data.

Extract useful features (day, weather, holidays).

Split into training/test sets.

Train a regression model.

Evaluate with MSE or MAE.

Deploy the model to help bakers plan production.

4. Failure Point + Fix

Failure: Bad features (e.g., ignoring holidays).
Fix: Add stronger features like holidays, weekends, or temperature to improve prediction
